```python
from PIL import Image
import numpy as np

# PNG 파일 읽기
image_path = "segmentation_mask.png"
image = Image.open(image_path)

# NumPy 배열로 변환 후 uint8로 변환
image_array = np.array(image, dtype=np.uint8)

# 확인
print("Array Shape:", image_array.shape)
print("Array Dtype:", image_array.dtype)
print("Unique Values:", np.unique(image_array))
```

```python
import matplotlib.pyplot as plt

# PNG 시각화
plt.imshow(image_array, cmap="tab10")  # 값이 0, 1, 2라면 색상으로 표현
plt.colorbar()  # 픽셀 값 범위 표시
plt.title("Visualization of PNG File")
plt.show()
```
